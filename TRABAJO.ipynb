```python
from IPython.display import IFrame  
FinalesLink="https://es.wikipedia.org/wiki/Anexo:Finales_de_la_Copa_Mundial_de_F%C3%BAtbol" 
IFrame(FinalesLink, width=900, height=900)
```

```python
import pandas as pd

FinalesDFs=pd.read_html(FinalesLink,header=0,flavor='bs4')
```

```python
#Averiguamos cuantas tablas hay en nuestra Data Frame
len(FinalesDFs)
```

```python
#Averiguamos el tipo de Data Frame que tenemos
type(FinalesDFs)
```

```python
#Realizamos una copia de nuestra Data Frame para trabajar con ella
Finales=FinalesDFs[0].copy()
```

```python
#Observamos nuestra copia
Finales
```

```python
#Averiguamos el tipo de data que tenemos en nuestra Data Frame
Finales.info()
```

```python
#Eliminamos la columna con la cual no queremos trabajar
NewFinales1=Finales.drop(columns='Notas')
```

```python
#Observamos nuestra nueva copia
NewFinales1
```

```python
#Eliminamos la fila con la cual no queremos trabajar
NewFinales2=NewFinales1.drop(22)
```

```python
#Observamos nuestra nueva copia,
#esta será nuestra Data Frame con la que trabajeremos
NewFinales2
```

```python
#Observamos el tipo de data que tenemos en nuestra Data Frame
NewFinales2.info()
```

```python
#Cambiamos el nombre de algunas columnas
NewFinales2 = NewFinales2.rename(columns={'Campeón': 'Ganador','Subcampeón': 'Perdedor','Espectadores': 'Asistentes','Localización': 'Ciudad'})
NewFinales2
```

```python
#Reemplazamos el nombre de Alemania Federal por Alemania
NewFinales2['Ganador'] = NewFinales2['Ganador'].str.replace('Alemania Federal', 'Alemania',regex=True)
NewFinales2
```

```python
#Eliminaremos los parentesis (y lo que encierra) de la columna Resultado
NewFinales2['Resultado'] = NewFinales2['Resultado'].str.replace(r'\([^()]*\)', '',regex=True)
NewFinales2
```

```python
#Observamos el tipo de data que tenemos en nuestra Data Frame
NewFinales2.info()
```

```python
#Eliminamos los espacios de las celdas al principio y al final,
#lo hacemos en todos ya que según el programa todos son de tipo object
NewFinales2['Fecha'] = NewFinales2['Fecha'].str.strip()
NewFinales2['Ganador'] = NewFinales2['Ganador'].str.strip()
NewFinales2['Resultado'] = NewFinales2['Resultado'].str.strip()
NewFinales2['Perdedor'] = NewFinales2['Perdedor'].str.strip()
NewFinales2['Estadio'] = NewFinales2['Estadio'].str.strip()
NewFinales2['Asistentes'] = NewFinales2['Asistentes'].str.strip()
NewFinales2['Ciudad'] = NewFinales2['Ciudad'].str.strip()
```

```python
NewFinales2
```

```python
#Ahora eliminamos los datos que no son números dentro de la columna Asistentes
NewFinales2['Asistentes']=NewFinales2.Asistentes.str.replace(pat=r'\D',repl='',regex=True)
```

```python
NewFinales2
```

```python
#De la columna Resultado, solo nos quedamos con el resultado del ganador
NewFinales2.Resultado=NewFinales2.Resultado.str.split(':',expand=True)[0]
```

```python
NewFinales2
```

```python
#Ahora eliminamos la columna de Perdedor
NewFinales2=NewFinales2.drop(columns='Perdedor')
NewFinales2
```

```python
#Cambiamos los nombres de las columnas
NewFinales2 = NewFinales2.rename(columns={'Ganador': 'Champion','Resultado': 'Goles','Estadio': 'Stadium','Asistentes': 'Assistants','Ciudad': 'City'})
NewFinales2
```

```python
NewFinales2.info()
```

```python
#Hallamos la cantidad de veces que 1 país ganó una copa del mundo
NewFinales2['Champion'].value_counts()
```

```python
#Hallamos la cantidad de veces que 1 ciudad fue sede de una final de un mundial
NewFinales2['City'].value_counts()
```

```python
#CÓDIGO ADICIONAL
#Observamos antes que la columna ¿Con cuánto ganó? y ¿Cuántas personas estuvieron? no son int o float,
#asi que cambiaremos para que sean datos enteros
NewFinales2['Goles'] = NewFinales2['Goles'].astype(int)
NewFinales2['Assistants'] = NewFinales2['Assistants'].astype(int)
```
